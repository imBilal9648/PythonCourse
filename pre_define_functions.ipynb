### Random Library

````python
import random
````

````python
random.random() # it will give random number between 0 to 1
````

````python
dice_value = random.randint(1,6)
if dice_value==6:
    print('you win')
else:
    print('try again, Loser ')    
````

````python
print('you win') if random.randint(1,6)==6 else print('try again, Loser')
````

````python
while True:
    dice_value = random.randint(1,6)
    if dice_value==6:
    print('you win')
    break
    else:
    print('try again, Loser! you got ',dice_value)
````

````python
emojis = ['😊','🤣','🤦‍♂️','🤷‍♀️','💖','🐱‍👤']
````

````python
print(random.choice(emojis))
````

````python
print(random.choices(emojis,k=3))
````
